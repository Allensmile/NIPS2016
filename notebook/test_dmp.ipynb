```python
import sys
sys.path.append('../')
import numpy as np
from nips.dmp.mydmp import MyDMP
import time
import matplotlib.pyplot as plt
%matplotlib inline

dmp = MyDMP()
```

```python
# RANDOM MOVEMENTS
for i in range(1):
    m = list(200. * (2. * np.random.random(7*4) - 1.)) + list(2.*np.random.random(4) -1.)
    m_dyn = dmp.trajectory(m)
    plt.plot(m_dyn)
    
plt.ylim([-1.1, 1.1])
```

```python
# IMITATE
m = list(300. * (2. * np.random.random(7*4) - 1.)) + list(2.*np.random.random(4) -1.)
m_dyn = dmp.trajectory(m)
plt.plot(m_dyn, "b", label="target")


t0 = time.time()
m = dmp.imitate(m_dyn, maxfun=1500)

print "time", time.time() - t0
plt.plot(dmp.trajectory(m), "r", label="optimized")
plt.ylim([-1.1, 1.1])
```
