```python
import sys
import numpy as np
sys.path.append('../')
from dmp.mydmp import DmpPrimitive
import time
import matplotlib.pyplot as plt
%matplotlib inline

dmp = DmpPrimitive()
```

```python
# RANDOM MOVEMENTS
for i in range(10):
    m = list(100. * (2. * np.random.random(7*4) - 1.)) + list(2.*np.random.random(4) -1.)
    m_dyn = dmp.trajectory(m)
    plt.plot(m_dyn)
    
plt.ylim([-1.1, 1.1])
```

```python
# IMITATE
m = list(300. * (2. * np.random.random(7*4) - 1.)) + list(2.*np.random.random(4) -1.)
m_dyn = dmp.trajectory(m)
plt.plot(m_dyn, "b", label="target")

#print m_dyn

t0 = time.time()
m = dmp.imitate(m_dyn, maxiter=50)

print "time", time.time() - t0
plt.plot(dmp.trajectory(m), "r", label="optimized")
plt.ylim([-1.1, 1.1])
```
